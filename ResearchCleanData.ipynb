```python
def check_Missing(dataset, precent):
  #Checking columns with missing data
  Miss_Percent=100*(dataset.isnull().sum()/len(dataset))

  #Sorting the data columns by their percentage in descending order (big to small)
  # by using Miss_Percent>0 it takes values,ore than 0
  # round(1) this is used to round the value to one decimal point .0
  Miss_Percent=Miss_Percent[Miss_Percent>precent].sort_values(ascending=False).round(1)

  #Creating a dataframe to show percentage of missing data and its respective data column in table
  DataFrame=pd.DataFrame(Miss_Percent)
  miss_percent_table=DataFrame.rename(columns={0:'% of Missing Values'})
  MissPercent=miss_percent_table

  #Displaying Missing Value table
  return MissPercent
```

```python
check_Missing(Data,50)
```

```python
import numpy as np
```

```python
def clean_data(Data,i='51'):
  #  To Remove nan values and replace blank 
  Data_New = Data.replace(np. nan,'',regex=True)
  Data_New['skills'] = Data_New.loc[:,'0':i].agg(' '.join, axis=1)
  Data_New.drop(Data_New.loc[:,'0':i],axis=1,inplace=True)
  Data_New['edu_1'] = Data_New[['education_degree_1','education_1']].agg(' '.join, axis=1)
  Data_New['edu_2'] = Data_New[['education_degree_2','education_2']].agg(' '.join, axis=1)
  Data_New['edu_3'] = Data_New[['education_degree_3','education_3']].agg(' '.join, axis=1)
  Data_New['org_1'] = Data_New[['organization_title_1','organization_1']].agg(' '.join, axis=1)
  Data_New['org_2'] = Data_New[['organization_title_2','organization_2']].agg(' '.join, axis=1)
  Data_New['org_3'] = Data_New[['organization_title_3','organization_3']].agg(' '.join, axis=1)
  Data_New['org_4'] = Data_New[['organization_title_4','organization_4']].agg(' '.join, axis=1)
  Data_New.drop(Data_New.loc[:,'education_1':'organization_start_4'],axis=1,inplace=True)

  return Data_New
```

```python
Data_set = clean_data(Data,'55')
```
