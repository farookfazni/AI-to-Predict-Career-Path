```python
import pandas as pd
```

```python
ML = pd.read_csv('/content/drive/MyDrive/Research Dataset/processed_data/Cleaned_Data_Step1/ML_processed1.csv')
DSci = pd.read_csv('/content/drive/MyDrive/Research Dataset/processed_data/Cleaned_Data_Step1/DSci_processed1.csv')
```

```python
ML_title = pd.read_csv('/content/drive/MyDrive/Research Dataset/processed_data/Fully_Extracted_Data/ML_processed1.csv')
DSci_title = pd.read_csv('/content/drive/MyDrive/Research Dataset/processed_data/Fully_Extracted_Data/DSci_processed1.csv')
```

```python
ML_title['organization_title_1']
```

```python
DSci_title['organization_title_1']
```

```python
ML_Data = pd.concat([ML,ML_title['organization_title_1']], axis=1)
DSci_Data = pd.concat([DSci,DSci_title['organization_title_1']], axis=1)
```

```python
Data = pd.concat([ML_Data,DSci_Data],ignore_index=True)
```

```python
# Drop Column
Data = Data.drop(['Unnamed: 0','Unnamed: 0.1'], axis=1)
```

```python
Data
```

```python
# Drop duplicate columns using a unique(key) Value
Data_new = Data.drop_duplicates(subset='id', keep="last")
```
