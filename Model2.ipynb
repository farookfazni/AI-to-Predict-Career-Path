```python
import spacy
import pandas as pd
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
Data = pd.read_csv('/content/drive/MyDrive/Research Dataset/processed_data/Added_Target/FullyMerged.csv')
```

```python
Data.head()
```

```python
Data = Data.drop(['Unnamed: 0'], axis=1)
```

```python
Data.head()
```

```python
# !pip install dframcy
# dframcy = DframCy(nlp)
# Data_doc = dframcy.nlp(Data)
```

```python
Data['ALL'] = Data[['skills','edu_1','edu_2','edu_3','org_1','org_2','org_3','org_4']].agg(' '.join, axis=1)
```

```python
Data_all = Data.iloc[:,9:]
Data_all
```

```python
Data_all['tokenized_ALL'] = Data_all['ALL'].apply(lambda x: nlp(x))
Data_all
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
label = le.fit_transform(Data_all['Target'])
# Data_tokenized.drop("Target", axis=1, inplace=True)
Data_all["Target_NO"] = label
Data_all
```

```python
token_data = Data_all.iloc[:,2:]
token_data
```

```python
Data_all['sent_ALL'] = Data_all['tokenized_ALL'].apply(lambda x: list(x.sents))
Data_all['sent_te_ALL'] = Data_all['sent_ALL'].apply(lambda x: [i.text for i in x])
Data_all['sent_te_ALL'] = Data_all['sent_ALL'].apply(lambda x: ' '.join(map(str,x)))
Data_all
```

```python
train, test = train_test_split(Data_all, test_size = 0.1, random_state = 17)
X_train = train['tokenized_ALL'].values
y_train = train['Target_NO']
 
X_test = test['tokenized_ALL'].values
y_test = test['Target_NO']
```

```python
from keras.models import Sequential
from keras.layers import Conv1D, GlobalMaxPooling1D, Embedding, LSTM
from keras.layers.core import Dense, Dropout, Activation, SpatialDropout1D
from keras import metrics, regularizers
```

```python
EMBEDDING_DIM = 200
```

```python
model = Sequential()

model.add(Embedding(4118,EMBEDDING_DIM,input_length=X_train.shape[0]))
model.add(SpatialDropout1D(0.2))
model.add(LSTM(250, dropout=0.2, return_sequences=True))


# model.add(Conv1D(nof_filters, kernel_size, padding='valid', activation='relu', strides = 1))
# model.add(GlobalMaxPooling1D())

model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
# model.add(Dropout(0.1))

# model.add(LSTM(32))
# model.add(Dropout(0.2))

model.add(Dense(1,activation='softmax'))

model.summary()
model.compile(loss='categorical_crossentropy', optimizer='adam', #'sgd', 'adam', 'RMSprop', 'Adagrad'
                   metrics = [metrics.categorical_accuracy])
```

```python
batch_size = 32
nb_epoch = 5
```

```python
history = model.fit(X_train, y_train,
                    batch_size = batch_size,
                    epochs = nb_epoch,
                    verbose = True,
                    validation_split = 0.2)
```
